# Memory Collection

The first step to live response collection is collecting memory. We will use `winpmem` to collect memory in this lab. To collect memory execute the following command:

```powershell
E:\winpmem.exe -m -p C:\pagefile.sys -o E:\memoryimage.aff
```

The `-p` flag specifies the location of the page file and the `-o` flag specfies the output location of the output of the collection. As you can see from the output of the command multiple files were collected. Explore the contents of the AFF file and see all what was collected. You can see the contents of a collected image with the `-V` flag.

```powershell
E:\winpmem.exe -V E:\memoryimage.aff
```

What is the name of the pagefile in the AFF image? Answer in the following cell in Markdown:

aff4://8bf786c8-8bfc-4000-93ea-28c6b26921e6/C:/pagefile.sys

The PMEM suite of acquistion tools always store the image file in AFF format. In the following labs we will be using volatility to perform analysis on the collected image but you could also use rekall if you prefer. To use volatility we have to extract the raw image from the AFF sparse image. To do this you have to specify the file to export. The following is an example command on how to perform this: 

`.\winpmem.exe memory.aff -e aff4://f8da0836-8871-4689-9580-d4a93ca66d87/PhysicalMemory -o memoryraw.dd`

Go ahead and extract the raw memory image from the one just collected for the next lab sections.
